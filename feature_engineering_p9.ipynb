```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, PowerTransformer
from sklearn.decomposition import PCA
```

```python
# Contoh Data
data = {"Luas Rumah (m2)": [120, 250, 85, 190],
        "Jumlah Kamar Tidur": [3, 5, 2, 4],
        "Lokasi": ["Kota A", "Kota B", "Kota C", "Kota D"],
        "Tahun Dibangun": [2000, 2010, 1995, 2020],
        "Harga Rumah": [300, 450, 200, 500]
        }
```

```python
# Dataframe
df = pd.DataFrame(data)
df
```

```python
# ---Feature Extraction---
# Memilih fitur-fitur penting
features = df[["Luas Rumah (m2)", "Jumlah Kamar Tidur", "Tahun Dibangun"]]
target = df["Harga Rumah"]
```

```python
# ---Feature Construction---
# 1. Menambahkan fitur baru Umur Rumah
# Create a new column for 'Umur Rumah' instead of overwriting 'Luas Rumah (m2)'
df["Umur Rumah"] = 2024 - df["Tahun Dibangun"]

# 2. Mengubah fitur kategorikal menjadi numerikal(one-hot encoding)
encoder = OneHotEncoder(sparse_output=False)  # or handle_unknown='ignore'
encoded_location = encoder.fit_transform(df[["Lokasi"]])
encoded_location_df = pd.DataFrame(encoded_location, columns=encoder.get_feature_names_out(["Lokasi"]))
```

```python
# Menambahka hasil encoding ke DataFrame
df = pd.concat([df, encoded_location_df], axis=1)

# Menghapus kolom yang tidak diperlukan
df.drop("Lokasi", axis=1, inplace=True)

# Hasil Akhir
print("Dataset setelah Feature Extraction dan Construction:")
df
```
