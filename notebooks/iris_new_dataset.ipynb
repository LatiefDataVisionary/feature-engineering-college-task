```python
import pandas as pd
import random
from google.colab import files
```

```python
# Dataset asli Iris untuk referensi
iris_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
# Fungsi untuk membuat dataset Iris baru
def create_custom_iris_dataset(num_samples=150, species_list=None, value_ranges=None):
    if species_list is None:
        species_list = ['setosa', 'versicolor', 'virginica']
    if value_ranges is None:
        value_ranges = {
            'sepal_length': (4.0, 8.0),
            'sepal_width': (2.0, 4.5),
            'petal_length': (1.0, 7.0),
            'petal_width': (0.1, 2.5)
        }

    # Generate data
    data = []
    for _ in range(num_samples):
        row = {
            'sepal_length': round(random.uniform(*value_ranges['sepal_length']), 2),
            'sepal_width': round(random.uniform(*value_ranges['sepal_width']), 2),
            'petal_length': round(random.uniform(*value_ranges['petal_length']), 2),
            'petal_width': round(random.uniform(*value_ranges['petal_width']), 2),
            'species': random.choice(species_list)
        }
        data.append(row)

    # Convert to DataFrame
    iris_df = pd.DataFrame(data, columns=iris_columns)
    return iris_df
```

```python
# Contoh penggunaan
custom_iris = create_custom_iris_dataset(
    num_samples=300000,
    species_list=['setosa', 'versicolor'],  # Hanya 2 spesies
    value_ranges={
        'sepal_length': (5.0, 7.5),
        'sepal_width': (2.5, 4.0),
        'petal_length': (1.5, 6.0),
        'petal_width': (0.5, 2.0)
    }
)
```

```python
# Simpan ke file CSV
custom_iris.to_csv('custom_iris_dataset.csv', index=False)
file_output = 'custom_iris_dataset.csv'
files.download(file_output)
```
